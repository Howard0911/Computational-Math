## Part A (1 point)

Describe, in words, what this code does, and how it does it.

````python
def f(x):
    if x == 0 or x == 1:
        return x
    return f(x - 1) + f(x - 2)
````

````python
for i in range(10):
    print(i,f(i))
````

The code describes Fibonacci sequence. The first term of the sequence is $0$, and the second term of the sequence is $1$. The other term of the sequence is the sum of previous two terms. The sequence is generated by recursion.

## Part B (2 points)

For what inputs will this function not behave as expected? What will happen?

````python
f(-1)
````

If the input is negative, this function will not behave as expected. Fibonacci sequence is only for integer. It will be in a infinite recursion.

````python
def f(x):
    assert x >= 0
    assert type(x) is int
    if x == 0 or x == 1:
        return x
    return f(x - 1) + f(x - 2)
````
